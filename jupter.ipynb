```python
import pandas as pd 
import numpy as np
import matplotlib as mp
import plotly.express as px
import seaborn as sb
```

```python
data = pd.read_csv("amazon_products.csv")
categories = pd.read_csv("amazon_categories.csv")
backup = data.copy()
print(data.info())
```

```python
data.nunique()
```

```python
data = backup
data.info()
```

```python
data = data.merge(categories, left_on='category_id', right_on='id', how='left')
data = data.rename(columns={'category_name': 'category'})
data = data.drop('id', axis=1)
data
```

```python
data = data.drop("imgUrl", axis = 1,  inplace = False)
data = data.drop("productURL", axis = 1,  inplace = False)
data = data.drop("asin", axis = 1, inplace = False)
data = data.drop("category_id", axis = 1, inplace=False)
data = data.dropna()
print(data.info())
```
